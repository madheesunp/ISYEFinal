```python
data = pd.read_csv('C:\\Users\\roadr\\GaTech\\Popular_Baby_Names.csv')
```

```python
data.head()
data["Child's First Name"] = data["Child's First Name"].str.upper()
```

```python
data['Ethnicity'] = data['Ethnicity'].replace({
    'ASIAN AND PACI': 'ASIAN AND PACIFIC ISLANDER',
    'BLACK NON HISP': 'BLACK NON HISPANIC',
    'WHITE NON HISP': 'WHITE NON HISPANIC'
})
```

```python
# Group the data by Ethnicity and Child's First Name, summing up the counts
aggregated_counts = data.groupby(['Ethnicity', "Child's First Name"])['Count'].sum().reset_index()

# Rename columns for clarity
aggregated_counts.columns = ['Ethnicity', "Child's First Name", 'Total Count']

# Sort the data for better readability
aggregated_counts = aggregated_counts.sort_values(by=['Ethnicity', 'Total Count'], ascending=[True, False])
```

```python
aggregated_counts
```

```python
# Calculate the total count for each ethnicity
ethnicity_total_counts = aggregated_counts.groupby('Ethnicity')['Total Count'].sum().reset_index()
ethnicity_total_counts.columns = ['Ethnicity', 'Ethnicity Total Count']

# Merge the total counts back into the most_popular_names dataframe
most_popular_names = most_popular_names.merge(ethnicity_total_counts, on='Ethnicity')

# Calculate the percentage for each ethnicity
total_count_all = ethnicity_total_counts['Ethnicity Total Count'].sum()
most_popular_names['Percentage of Total'] = (most_popular_names['Ethnicity Total Count'] / total_count_all) * 100

# Drop the ethnicity total count column for cleaner presentation
most_popular_names = most_popular_names.drop(columns=['Ethnicity Total Count'])
```

```python
# Group the data by year, ethnicity, and name to calculate the total count for each combination
yearly_counts = data.groupby(['Year of Birth', 'Ethnicity', "Child's First Name"])['Count'].sum().reset_index()

# Calculate the total count for each year and ethnicity
total_count_by_year_ethnicity = yearly_counts.groupby(['Year of Birth', 'Ethnicity'])['Count'].sum().reset_index()
total_count_by_year_ethnicity.columns = ['Year of Birth', 'Ethnicity', 'Total Count']

# Find the most popular and least popular names for each year and ethnicity
most_popular_names_by_year = yearly_counts.loc[yearly_counts.groupby(['Year of Birth', 'Ethnicity'])['Count'].idxmax()]
least_popular_names_by_year = yearly_counts.loc[yearly_counts.groupby(['Year of Birth', 'Ethnicity'])['Count'].idxmin()]

# Merge the most popular, least popular, and total count data
most_popular_names_by_year = most_popular_names_by_year.rename(columns={
    "Child's First Name": "Most Popular Name",
    "Count": "Most Popular Count"
})
least_popular_names_by_year = least_popular_names_by_year.rename(columns={
    "Child's First Name": "Least Popular Name",
    "Count": "Least Popular Count"
})

# Merge the datasets
summary_by_year_ethnicity = total_count_by_year_ethnicity.merge(
    most_popular_names_by_year[['Year of Birth', 'Ethnicity', 'Most Popular Name', 'Most Popular Count']],
    on=['Year of Birth', 'Ethnicity']
).merge(
    least_popular_names_by_year[['Year of Birth', 'Ethnicity', 'Least Popular Name', 'Least Popular Count']],
    on=['Year of Birth', 'Ethnicity']
)
```
