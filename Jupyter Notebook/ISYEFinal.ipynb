```python
import pandas as pd
import matplotlib.pyplot as plt
import dataframe_image as dfi
```

```python
data = pd.read_csv('C:\\Users\\roadr\\GaTech\\Popular_Baby_Names.csv')
```

```python
data.head()
data["Child's First Name"] = data["Child's First Name"].str.upper()
```

```python
data['Ethnicity'] = data['Ethnicity'].replace({
    'ASIAN AND PACI': 'ASIAN AND PACIFIC ISLANDER',
    'BLACK NON HISP': 'BLACK NON HISPANIC',
    'WHITE NON HISP': 'WHITE NON HISPANIC'
})
# data = data.replace('ASIAN AND PACIFIC ISLANDER', 'ASIAN AND PACI')
# data = data.replace('WHITE NON HISPANIC', 'WHITE NON HISP')
```

```python
most_popular_names = data.loc[data.groupby(['Ethnicity', "Child's First Name"])['Count'].count(), ['Ethnicity', "Child's First Name", 'Count']]

most_popular_names
```

```python
most_popular_name_byYear = data.loc[data.groupby(['Year of Birth', 'Ethnicity', 'Gender'])['Count'].idxmax(), ['Ethnicity', "Child's First Name", 'Count', 'Year of Birth', 'Gender']]

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', 0)

most_popular_name_byYear
```

```python
# Group the data by Ethnicity and calculate the variance of the Count column
variance_by_ethnicity = data.groupby(["Ethnicity","Child's First Name")['Count'].var()
```

```python
# Convert the result to a DataFrame for better readability
variance_by_ethnicity_df = variance_by_ethnicity.reset_index()
variance_by_ethnicity_df.columns = ['Ethnicity', 'Variance']

# Display the results
variance_by_ethnicity_df
```
